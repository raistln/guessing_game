```python
def bigger_lesser(num_range, n=2):
    choice = choices(num_range[1:-1])[0]
    if n % 2 == 0:
        bigger = input(f"It´s your number bigger as {choice}. (Answer only y or n)")
        if bigger == "y":
            num_range = range(choice, num_range[-1]+1)
        elif bigger == "n":
            num_range = range(num_range[0], choice+1)
        else:
            print("Let´s try again.")
            return bigger_lesser(num_range, n=n)
    if n % 2 == 1:
        lesser = input(f"It´s your number lesser as {choice}. (Answer only y or n)")
        if lesser == "y":
            num_range = range(num_range[0], choice+1)
        elif lesser == "n":
            num_range = range(choice, num_range[-1]+1)
        else:
            print("Let´s try again.")
            return bigger_lesser(num_range, n=n)
    n += 1
    
    if len(num_range) == 1:
        return print("Oh, my dear I WON.")
    elif len(num_range) == 2:
        final_choice = choices(num_range, k=2)
        answer = input(f"Your number is {final_choice[0]}?")
        if answer == "y":
            return print("Oh, my dear I WON.")
        elif answer == "n":
            return print(f"Shit I loose your number is {final_choice[1]}.")
    return bigger_lesser(num_range, n=n)
```

```python
def begin_game(number):
    if type(number) != int and number < 0:
        print("Don´t try to fool me, thast not a number and you know it.")
    else:
        try:
            num_range = range(number+1)
            bigger_lesser(num_range)
        except ValueError:
            print("Ok, ok, ok. Something went wrong, let´s try from the beggining")
            begin_game(number)
```
